```python
import pandas as pd
```

```python
file_path = 'DaNang_2000_2024.csv'
data = pd.read_csv(file_path)
```

```python
data.head(5)
```

```python
data.describe()
```

```python
data['dt_iso'] = pd.to_datetime(data['dt_iso'].str.replace(' +0000 UTC', '', regex=False))
```

```python
data.isnull().sum
```

```python
data['rainfall'] = data['rain_1h']
data['month'] = data['dt_iso'].dt.to_period('M')
selected_columns = ['month', 'temp', 'humidity', 'wind_speed', 'rainfall', 'clouds_all']
data_selected = data[selected_columns]
```

```python
data_monthly = data_selected.groupby('month').mean().reset_index()
```

```python
data_monthly.head()
```

```python

data_monthly.to_csv('montlydata.csv', index=False)
```

```python
data['year'] = data['dt_iso'].dt.year
data['month'] = data['dt_iso'].dt.month
```

```python
monthly_rainfall = data.groupby(['year', 'month'])['rain_1h'].mean().reset_index()
```

```python
monthly_rainfall.rename(columns={'rain_1h': 'rainfall_data'}, inplace=True)
```

```python
annual_rainfall_data = monthly_rainfall.pivot(index='year', columns='month', values='rainfall_data').fillna(0)
annual_rainfall_data.head(5)
```

```python
annual_rainfall_data['annual_rainfall'] = annual_rainfall_data.sum(axis=1)
```

```python
annual_rainfall_data.columns = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANNUAL RAINFALL']
```

```python
annual_rainfall_data
```

```python
annual_rainfall_data.drop(annual_rainfall_data.index[-1])
```
